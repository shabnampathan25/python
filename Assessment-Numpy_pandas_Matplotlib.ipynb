### Q 1: Develop a Line chart using the functionality of pandas to show how automobile sales fluctuate from year to year. 

```python
import matplotlib.pyplot as plt

df1 = pd.read_csv("C://Users//Acer//Downloads//historical_automobile_sales.csv")
df1
```

```python
yearly_sales = df1.groupby("Year")["Automobile_Sales"].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(yearly_sales["Year"], yearly_sales["Automobile_Sales"], marker="o", linestyle="-", color="b")

plt.title("Yearly Automobile Sales Fluctuation", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Automobile Sales", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.xticks(yearly_sales["Year"], rotation=45)
plt.tight_layout()
plt.show()
```

### Q 2: Plot different lines for categories of vehicle type and analyze the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods? 

```python
recession_data = df1[df1["Recession"] == 1]

sales_by_vehicle_type = (
    recession_data.groupby(["Year", "Vehicle_Type"])["Automobile_Sales"]
    .sum()
    .reset_index()
)

sales_pivot = sales_by_vehicle_type.pivot(index="Year", columns="Vehicle_Type", values="Automobile_Sales")

plt.figure(figsize=(12, 8))
for vehicle_type in sales_pivot.columns:
    plt.plot(sales_pivot.index, sales_pivot[vehicle_type], marker="o", label=vehicle_type)

plt.title("Automobile Sales Trends by Vehicle Type During Recession Periods", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Automobile Sales", fontsize=12)
plt.legend(title="Vehicle Type")
plt.grid(True, linestyle="--", alpha=0.6)
plt.xticks(sales_pivot.index, rotation=45)
plt.tight_layout()
plt.show()
```

```python
#or
sales_pivot = (
    df1[df1["Recession"] == 1]
    .groupby(["Year", "Vehicle_Type"])["Automobile_Sales"]
    .sum()
    .unstack("Vehicle_Type")
)

sales_pivot.plot(figsize=(12, 8), marker="o")

plt.title("Automobile Sales Trends by Vehicle Type During Recession Periods", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Automobile Sales", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(title="Vehicle Type")
plt.tight_layout()
plt.show()
```

### Q 3: Use the functionality of Seaborn Library to create a visualization to comparethe sales trend per vehicle type for a recession period with a non- recession period. 

```python
import seaborn as sns

df1['Date'] = pd.to_datetime(df1['Date'])
df1['Recession_Period'] = df1['Recession'].apply(lambda x: 'Recession' if x == 1 else 'Non-Recession')
plt.figure(figsize=(14, 8))
sns.lineplot(x='Date', y='Automobile_Sales', hue='Vehicle_Type', style='Recession_Period', markers=True, data=df1)
plt.title('Sales Trend per Vehicle Type: Recession vs Non-Recession Period')
plt.xlabel('Date')
plt.ylabel('Automobile Sales')
plt.xticks(rotation=45)
plt.legend(title='Vehicle Type & Period')
plt.tight_layout()
plt.show()
```

### Q 4: Now you want to compare the sales of different vehicle types during a recession and a non-recession period 
Hint: To visualize sales of different vehicles during recession and non- recession periods, you
can use a bar chart. You will need to group Recession, Vehicle_Typefor average
Automobile_Sales and then plot it. Make use of sns. bar plot (x=x, y=data = df)) 


```python
grouped_data = df1.groupby(['Recession', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# Creating a bar plot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Vehicle_Type', y='Automobile_Sales', hue='Recession', data=grouped_data)

# Adding labels and title
plt.xlabel('Vehicle Type')
plt.ylabel('Average Automobile Sales')
plt.title('Comparison of Vehicle Sales During Recession and Non-Recession Periods')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
